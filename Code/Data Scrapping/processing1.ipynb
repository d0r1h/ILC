```python
import os
import requests
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
path = "../../Data/Indian Legal Cases/"
```

```python
briefcased = pd.read_csv(path + "briefcased.csv")
primelegal = pd.read_csv(path + "primelegal.csv")
```

```python
DF = pd.concat([briefcased, primelegal])
DF.reset_index(inplace=True, drop=True)
```

```python
DF.dropna(inplace=True)
DF.reset_index(inplace=True,drop=True)
```

```python
DF[DF.duplicated()]
```

```python
for i, j in enumerate(DF['titles']):
    if "Death penalty " in j:
        print(i, j)
```

```python
DF.iloc[[624, 625]]
```

```python
DF.drop_duplicates(keep='first', inplace=True)
DF.reset_index(inplace=True, drop=True)
```

```python
DF[DF.duplicated()]
```

```python
DF.isna().sum()
```

```python
corrupt = []

for i,j in enumerate(DF['links']):
    if ("primelegal" not in j) and ("briefcased" not in j) and ("bombayhighcourt" not in j) \
        and ("164.100.69.66" not in j) and ("main.sci.gov.in" not in j) and ("livelaw" not in j) \
        and ("mhc.tn.gov.in" not in j) and ("karnatakajudiciary.kar.nic.in" not in j) \
        and ("highcourtofuttarakhand.kar.nic.in" not in j)and ("nclat.nic.in" not in j) \
        and ("theleaflet" not in j) and ("cmshck.kerala.gov.in" not in j): 
        
        corrupt.append(i)
        
corrupt.append(3620) # There is one .php file in karnatakajudiciary and it is corrupted
```

```python
len(corrupt)
```

```python
DF.iloc[corrupt]
```

```python
DF.drop(corrupt, inplace=True)
DF.reset_index(inplace=True,drop=True)
```

```python
try:
    %mkdir Cases Summaries Titles
except Exception as e:
    pass
```

**Judgment Cases**

```python
for i, link in enumerate(DF['links']):
    path = "Cases/"
    try:
        response = requests.get(link)
        with open(path+"Case ID"+str(i)+".pdf", "wb") as f:
            f.write(response.content)
    except Exception as e:
        try:
            response = requests.get(link, verify=False)
            with open(path+"Case ID"+str(i)+".pdf", "wb") as f:
                f.write(response.content)
        except Exception as e:
            print(e)
```

**Judgment Summaries**

```python
for i, summary in enumerate(DF['summaries']):
    path = "Summaries/"
    with open(path+"Sum ID"+str(i)+".txt", "wb") as f:
        f.write(summary.encode("utf-8"))
```

**Judgment Titles**

```python
for i, title in enumerate(DF['titles']):
    path = "Titles/"
    with open(path+"Title ID"+str(i)+".txt", "wb") as f:
        f.write(title.encode("utf-8"))
```
