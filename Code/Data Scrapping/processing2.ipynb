                                        From File to DataFrame & Json Files

```python
import json
import os
import pandas as pd
```

```python
titles = "../../Data/Indian Legal Cases/ILC - v.1.1/Titles/"
summaries = "../../Data/Indian Legal Cases/ILC - v.1.1/Summaries/"
cases = "../../Data/Indian Legal Cases/ILC - v.1.1/Cases/"
```

```python
with open(titles + 'Title ID10.txt') as f:
    text = f.read()
print(text)
```

```python
with open(summaries + 'Sum ID10.txt', encoding='utf-8') as f:
    text = f.read()
print(text)
```

```python
with open(cases + 'Case ID10.txt', encoding='utf-8') as f:
    text = f.read()
print(text)
```

**DataFrame**

```python
Title, Summaries, Cases = [], [], []

for i in range(len(os.listdir(cases))):

    with open(titles + f'Title ID{i}.txt', encoding='utf-8') as f:
        Title.append(f.read())

    with open(summaries + f'Sum ID{i}.txt', encoding='utf-8') as f:
        Summaries.append(f.read())

    with open(cases + f'Case ID{i}.txt', encoding='utf-8') as f:
        Cases.append(f.read())
```

```python
DF = pd.DataFrame(list(zip(Title, Summaries, Cases)), columns =['Title', 'Summary', 'Case'])
```
